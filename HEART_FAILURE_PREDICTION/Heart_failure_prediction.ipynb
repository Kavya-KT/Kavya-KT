```python
X=df.iloc[:,:-1]
X
```

```python
y=df.iloc[:,-1]
y
```

```python
from sklearn.preprocessing import MinMaxScaler
m=MinMaxScaler()
X_m=m.fit_transform(X)

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_m,y,random_state=1,test_size=0.3)

# print(X_train.shape,X_test.shape)
# print(y_train.shape,y_test.shape)


from sklearn.neighbors import KNeighborsClassifier
k=KNeighborsClassifier()
k.fit(X_train,y_train)
y_k=k.predict(X_test)

from sklearn.metrics import ConfusionMatrixDisplay,classification_report
print(classification_report(y_test,y_k))
print(ConfusionMatrixDisplay.from_predictions(y_test,y_k))
```

# **hypertuning parametre**

```python
from sklearn.model_selection import GridSearchCV
p={'n_neighbors':[1,3,5,7,9],'weights':['uniform','distance'],'algorithm':['auto','ball_tree','kd_tree','brute']}
t=GridSearchCV(k,p,cv=5,scoring='accuracy')
t.fit(X_train,y_train)
print(t.best_params_)
```

```python
kt=KNeighborsClassifier(algorithm= 'auto', n_neighbors= 5, weights= 'distance')
kt.fit(X_train,y_train)
y_kt=kt.predict(X_test)


print(classification_report(y_test,y_kt))
print(ConfusionMatrixDisplay.from_predictions(y_test,y_kt))
```
